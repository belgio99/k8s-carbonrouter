```python
import json
import matplotlib.pyplot as plt

# Load the scenario data
with open('carbon_scenario.json', 'r') as f:
    scenario_data = json.load(f)

pattern = scenario_data['pattern']
description = scenario_data['description']

# Calculate time points (15 seconds per point)
time_points = [i * 15 for i in range(len(pattern))]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(time_points, pattern, marker='o', linestyle='-', color='skyblue')
plt.title(f'Carbon Intensity Scenario: {description}')
plt.xlabel('Time (seconds)')
plt.ylabel('Carbon Intensity (gCO2/kWh)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
